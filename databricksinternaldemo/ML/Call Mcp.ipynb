```python
from databricks.sdk import WorkspaceClient
from databricks_mcp import DatabricksMCPClient

workspace_client = WorkspaceClient()
host = workspace_client.config.host

VECTOR_SERVER_URL = f"{host}/api/2.0/mcp/vector-search/na-dbxtraining/biju_raw"

mcp_client = DatabricksMCPClient(
    server_url=VECTOR_SERVER_URL,
    workspace_client=workspace_client
)

tools = mcp_client.list_tools()
display([t.name for t in tools])

response = await mcp_client.call_tool_async(
    "na_dbxtraining__biju_raw__ticket_embeddingsindex",
    arguments={"query": "How to reset password?"}
)

for content in response.content:
    display(content.text)
```
