```python
%sql
SELECT 
    pickup_date,
    vendor_id,
    daily_revenue,
    daily_trips,
    SUM(daily_revenue) OVER (
        PARTITION BY vendor_id 
        ORDER BY pickup_date 
        ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
    ) as running_total_revenue,
    AVG(daily_revenue) OVER (
        PARTITION BY vendor_id 
        ORDER BY pickup_date 
        ROWS BETWEEN 6 PRECEDING AND CURRENT ROW
    ) as moving_avg_7day
FROM (
    SELECT 
        to_date(f.lpep_pickup_datetime) as pickup_date,
        f.VendorID as vendor_id,
        SUM(f.total_amount) as daily_revenue,
        COUNT(*) as daily_trips
    FROM `na-dbxtraining`.biju_silver.fact_green_trips f
    WHERE to_date(f.lpep_pickup_datetime) >= '2024-01-01' 
        AND to_date(f.lpep_pickup_datetime) < '2024-02-01'
    GROUP BY to_date(f.lpep_pickup_datetime), f.VendorID
)
ORDER BY vendor_id, pickup_date
```
